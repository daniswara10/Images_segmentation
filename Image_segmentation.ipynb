```python
from skimage.io import imread, imshow
from skimage.color import rgb2hsv
import matplotlib.pyplot as plt

# Membaca gambar
image = imread('/content/drive/MyDrive/Colab Notebooks/bags.png')  # Pastikan path benar
hsv_image = rgb2hsv(image)

# Membuat subplot
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

# Menampilkan komponen Hue
ax[0].imshow(hsv_image[:, :, 0], cmap='gray')
ax[0].set_title('Hue')

# Menampilkan komponen Saturation
ax[1].imshow(hsv_image[:, :, 1], cmap='gray')
ax[1].set_title('Saturation')

# Menampilkan komponen Value
ax[2].imshow(hsv_image[:, :, 2], cmap='gray')
ax[2].set_title('Value')

plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(hsv_image[:,:,0],cmap='hsv')
ax[0].set_title('hue')
ax[1].imshow(hsv_image[:,:,1],cmap='hsv')
ax[1].set_title('transparency')
ax[2].imshow(hsv_image[:,:,2],cmap='hsv')
ax[2].set_title('value')
fig.colorbar(imshow(hsv_image[:,:,0],cmap='hsv'))
fig.tight_layout()
```

```python
import numpy as np  # Mengimpor numpy

# Mengacu pada channel hue (di colorbar)
lower_mask = hsv_image[:, :, 0] > 0.6  # Bagian dari hue channel
upper_mask = hsv_image[:, :, 0] < 0.7  # Bagian dari hue channel

# Mengacu pada channel saturation (di colorbar)
saturation_mask = hsv_image[:, :, 1] > 0.3  # Bagian dari saturation channel

# Menggabungkan semua mask
mask = upper_mask * lower_mask * saturation_mask

# Mengaplikasikan mask ke masing-masing channel RGB
red = image[:, :, 0] * mask
green = image[:, :, 1] * mask
blue = image[:, :, 2] * mask

# Menggabungkan channel RGB yang telah di-mask
bags_masked = np.dstack((red, green, blue))

# Menampilkan gambar hasil
imshow(bags_masked)
plt.show()
```
